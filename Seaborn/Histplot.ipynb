Common_norm : If True and using a normalized statistic, the normalization will apply over the full dataset. 

Otherwise, normalize each histogram independently.



```python
sns.histplot(
    penguins, x="bill_length_mm", hue="island", element="step",
    stat="density", common_norm=True,
)
```

```python
sns.histplot(
    penguins, x="bill_length_mm", hue="island", element="step",
    stat="density", common_norm=False,
)
```

discrete  : If True, default to binwidth=1 and draw the bars so that they are centered on their corresponding data points. 

This avoids “gaps” that may otherwise appear when using discrete (integer) data.



```python
tips = sns.load_dataset("tips")
sns.histplot(data=tips, x="size", stat="percent", discrete=True)
```

```python
tips = sns.load_dataset("tips")
sns.histplot(data=tips, x="size", stat="percent",  discrete=False)
```

Shrink : Scale the width of each bar relative to the binwidth by this factor. Only relevant with univariate data.

You can even draw a histogram over categorical variables (although this is an experimental feature)

```python
sns.histplot(data=tips, x="day", shrink=.9)
```

```python
sns.histplot(data=tips, x="day", hue="sex", shrink=.8, stat='percent', multiple='s')
```

# Real-world data is often skewed. For heavily skewed distributions, it’s better to define the bins in log space

```python
planets = sns.load_dataset("planets")
sns.histplot(data=planets, x="distance")
```

```python
sns.histplot(data=planets, x="distance", log_scale=True)
```

There are also a number of options for how the histogram appears. You can show unfilled bars

```python
sns.histplot(data=planets, x="distance", log_scale=True, fill=False)
```

```python
sns.histplot(data=planets, x="distance", log_scale=True, element="step", fill=False)
```
