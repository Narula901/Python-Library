```python
import seaborn as sns
```

```python
tips = sns.load_dataset("tips")
sns.kdeplot(data=tips, x="total_bill")
```

Seaborn allows you to customize the smoothing (or bin width) or a kernel density estimate by using the bw_adjust= parameter. 

The large the value, the smoother the curve. We can add more detail to the curve by lowering the value.

```python
tips
```

```python
ax= sns.kdeplot(data=tips, x="total_bill", bw_adjust=5)
```

bw_adjust

we passed in bw_adjust=0.3. This reduces the smoothing and adds more detail in the fluctuations of the underlying function, as shown below:

```python
ax= sns.kdeplot(data=tips, x="total_bill", bw_adjust=0.3)
```

cut

Factor, multiplied by the smoothing bandwidth, that determines how far the evaluation grid extends past the extreme datapoints. When set to 0, truncate the curve at the data limits.

Use more smoothing, but donâ€™t smooth past the extreme data points:



```python
ax= sns.kdeplot(data=tips, x="total_bill", bw_adjust=5, cut=0)
```

```python
ax= sns.kdeplot(data=tips, x="total_bill", bw_adjust=5, cut=0.8)
```

```python
ax= sns.kdeplot(data=tips, x="total_bill", bw_adjust=5, cut=1)
```

```python
ax= sns.kdeplot(data=tips, x="total_bill", bw_adjust=5, cut=0.4)
```

```python
ax= sns.kdeplot(data=tips, x="total_bill", bw_adjust=5, cut=5)
```

```python
ax= sns.kdeplot(data=tips, x="total_bill", bw_adjust=5, cut=10)
```

```python
sns.kdeplot(data=tips, x="total_bill", hue="time")
```

# Possible values of Multiple ----- ['layer', 'stack', 'fill']

```python
sns.kdeplot(data=tips, x="total_bill", hue="time", multiple="stack")
```

```python
sns.kdeplot(data=tips, x="total_bill", hue="time", multiple="fill")
```

Normalize the stacked distribution at each value in the grid

```python
sns.kdeplot(data=tips, x="total_bill", hue="time", multiple="layer")
```

```python
sns.kdeplot(data=tips, x="total_bill", hue="time", multiple="layer", common_norm=False)
```

```python
sns.kdeplot(data=tips, x="total_bill", hue="time", multiple="layer", common_norm=True)
```

Estimate the cumulative distribution function(s), normalizing each subset:

```python
sns.kdeplot(data = tips, x='total_bill', hue='time', cumulative=True, common_norm = False, common_grid=True)
```

Common_norm --- If True, scale each conditional density by the number of observations such that the total area under all densities sums to 1.

Otherwise, normalize each density independently

```python
sns.kdeplot(
   data=tips, x="total_bill", hue="size",
   fill=True, common_norm=False, palette="crest",
   alpha=.5, linewidth=0,
)
```

```python
sns.kdeplot(data=tips, x="total_bill", hue="size", fill=True, common_norm=True, palette="crest", alpha=.5, linewidth=0)
```
