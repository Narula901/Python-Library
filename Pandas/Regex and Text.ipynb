### 265. Slicing Substrings

    The syntax for using slice is slice(start, stop, step).

```python
s
```

```python
s[slice(0,7,1)]
```

```python
s[slice(7)]
```

```python
s[0:7:1]
```

```python
s[:7:]
```

```python
Boston_Marathon.head()
```

```python
Boston_Marathon.Country.str.slice(0,2,1)
```

    start=-2: This specifies the starting position for the slice. When set to -2, it means to start from the second-to-last character in each string.

    step=1: The step determines how the slicing progresses through the string. A step of 1 means that it takes every character.

    stop=None: The stop parameter is set to None, which means it will slice until the end of the string.

```python
Boston_Marathon['Country']
```

```python
Boston_Marathon.Country.str.slice(start = -2, step = 1, stop = None)
```
