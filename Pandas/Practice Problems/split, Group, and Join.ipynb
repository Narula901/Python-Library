```python
df['Address'] = df['Purchase Address'].apply(lambda x: x.split(',')[1].strip() if len(x.split(',')) > 1 else '')
```

```python
most_sold_product = Sales_data.groupby('Product').agg({'Quantity Ordered':'count'}).sort_values( by = 'Quantity Ordered', ascending = False).nlargest(5, 'Quantity Ordered').index
```

```python
df[df['Product'].isin(most_sold_product)]
```

```python
Lowest_sold_product = Sales_data.groupby('Product').agg({'Quantity Ordered':'count'}).sort_values( by = 'Quantity Ordered', ascending = False).nsmallest(5, 'Quantity Ordered').index
```
