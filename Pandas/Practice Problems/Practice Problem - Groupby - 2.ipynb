```python
import pandas as pd

# Initialize an empty list to store DataFrames
dfs = []

# Iterate over the Genre groups
for genre, data in Genre:
    # Find the row with the highest global sales for each genre
    Highest_Global_Sales_in_group = data.nlargest(1, 'Global_Sales')
    # Append this DataFrame to the list
    dfs.append(Highest_Global_Sales_in_group)

# Concatenate all DataFrames in the list to create the final DataFrame
df = pd.concat(dfs, ignore_index=True)

# Now df contains the concatenated data
```
