```python
midterms = pd.DataFrame(
    [['alex', 'padilla', 92], ['rayna', 'wilson', 83], ['juan', 'gomez', 78], ['angela', 'smith', 66],['stephen', 'yu', 98]], 
    columns=['first', 'last', 'score']
)
finals = pd.DataFrame(
    [['alex','padilla', 97, False], ['rayna', 'wilson', 88, False], ['alex', 'smith', 86, True], ['juan', 'gomez', 71, True], ['stephen', 'yu', 78, False], ['sakura', 'steel', 100, True]], 
    columns=['first', 'last','score', 'extra_credit']
)
```

```python
midterms
```

```python
finals
```

```python
combine = midterms.merge(finals, how = 'inner', on = ['first', 'last'], suffixes= ['_midterm', '_final'])
```

```python
combine.head()
```

```python
combine['Avg'] = combine.apply(lambda x : (x['score_midterm'] + x['score_final'])/2, axis =1)
```

```python
combine
```

```python
combine.loc[combine['extra_credit'] == True, 'Avg'] +=10
```

```python
combine
```

```python
combine.loc[combine['extra_credit'] == False, 'Avg'] +=2
```
