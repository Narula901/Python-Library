```python
def check_for_indpendence(df,col1,col2):
    col1_col2 = df[[col1, col2]]
    col1_col2_CT = col1_col2.pivot_table(index = col1_col2[col1], columns = col1_col2[col2], aggfunc = len)
    col1_col2_CT.replace(np.nan, 0, inplace=True)
    chi2, p, dof, tbl = chi2_contingency(col1_col2_CT) 

    print(f"Chi-squared statistic: {chi2}")
    print(f"P-value: {p}")
    print(f"dof: {dof}")
    print(f"tbl: {tbl}")

    print()

    if p < 0.05:
        print(f"There is a relationship between {col1} and {col2}.")
    else:
        print(f"No relationship exists between {col1} and {col2}.")

    
```
