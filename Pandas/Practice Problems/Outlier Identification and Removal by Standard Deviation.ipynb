# Outlier Identification and Removal

```python
import pandas as pd
import numpy as np
```

```python
games = pd.read_csv('D:\Data for Preprocessing\games_sales.csv')
```

```python
numerical= games.select_dtypes(include=['float64'])
```

```python
numerical.describe()
```

```python
# Calculate mean and standard deviation for each column
means = numerical.mean()
std_devs = numerical.std()

# Set a threshold (e.g., 2 times the standard deviation for each column)
thresholds = 3 * std_devs

# Identify outliers across all columns
outliers = numerical[(np.abs(numerical - means) > thresholds).any(axis=1)]

print("Means:")
print(means)
print("\nStandard Deviations:")
print(std_devs)
print("\nThresholds:")
print(thresholds)
print("\nOutliers:")
print(outliers)
```

```python
# Remove outliers from the main DataFrame
data_without_Outlier = numerical[~(np.abs(numerical - means)> thresholds).any(axis=1)]
```
