```python
import pandas as pd
import numpy as np
```

```python
transactions = pd.read_csv('D:\\karan data\\Data for Preprocessing\\retail\\transactions.csv')
transactions
```

```python
transactions['date']= pd.to_datetime(transactions['date'], errors='coerce')
```

```python
transactions = transactions.assign (
    pct_to_target = transactions['transactions']/2500,
    met_target = transactions['transactions'] >= 1,
    bonus_payable = (transactions['transactions'] >= 1) * 100,
    Month = transactions['date'].dt.month,
    day_of_week = transactions['date'].dt.day_of_week
)
```

```python
transactions
```

```python
conditions = [
    transactions['Month'] == 12,
    (transactions['Month'] == 5) & (transactions['day_of_week'] == 6),
    (transactions['Month'] == 7) & (transactions['day_of_week'] == 0)
]

choices = ['Holiday Bonus', 'Corporate Month', 'Summer Special']

transactions['seasonal_bonus'] = np.select(conditions, choices, default='None')
```

```python
transactions[transactions['seasonal_bonus'] == 'Summer Special']
```

```python
import numpy as np

# Assuming transactions is your DataFrame containing columns 'Month' and 'day_of_week'

def seasonal_bonus(month, day_of_week):
    if month == 12:
        return 'Holiday Bonus'
    elif month == 5 and day_of_week == 6:
        return 'Corporate Month'
    elif month == 7 and day_of_week == 0:
        return 'Summer Special'
    else:
        return 'None'

transactions['seasonal_bonus_1'] = transactions.apply(lambda x: seasonal_bonus(x['Month'], x['day_of_week']), axis=1)
```
