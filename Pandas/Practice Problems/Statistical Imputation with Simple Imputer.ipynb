```python
import pandas as pd
```

```python
Kidney_disease = pd.read_csv("D:\Data for Preprocessing\kidney_disease.csv")
```

```python
Kidney_disease.head()
```

```python
Kidney_disease.info()
```

```python
Kidney_disease['Packed Cell Volume'] = pd.to_numeric(Kidney_disease['Packed Cell Volume'], errors='coerce')
Kidney_disease['White Blood Cell Count'] = pd.to_numeric(Kidney_disease['White Blood Cell Count'], errors='coerce')
Kidney_disease['Red blood cell count'] = pd.to_numeric(Kidney_disease['Red blood cell count'], errors='coerce')

# Now converting to float64
Kidney_disease = Kidney_disease.astype({'Packed Cell Volume': 'float64', 'White Blood Cell Count': 'float64', 'Red blood cell count': 'float64'})
```

```python
Kidney_disease.info()
```

```python
Categorical_data = Kidney_disease.select_dtypes(include = 'object')
```

```python
Categorical_data
```

```python
Numerical_data = Kidney_disease.select_dtypes(include = 'float64')
```

```python
Numerical_data
```

```python
Numerical_data.isnull().sum()
```

```python
from sklearn.impute import SimpleImputer
```

```python
# Imputation with Mean Value
imputer = SimpleImputer(strategy = 'mean')  ### instead of mean, we can use median, mode and constant value
```

```python
imputer.fit(Numerical_data)
```

```python
Imputed_Data_mean = pd.DataFrame(imputer.transform(Numerical_data))
```

```python
Imputed_Data_mean.columns = Numerical_data.columns
```

```python
# Impute Numerical value with mean 
Imputed_Data_mean
```

```python
# Others way to fill nan values
Categorical_data.fillna(method = 'bfill') #ffill
```

```python
Categorical_data.isnull().sum()
```

```python
imputer_cat = SimpleImputer(strategy = 'most_frequent')
```

```python
imputer_cat.fit(Categorical_data)
```

```python
Imputed_Data_mode = pd.DataFrame(imputer_cat.transform(Categorical_data))
```

```python
Imputed_Data_mode.isnull().sum()
```

```python
Imputed_Data_mode.columns = Categorical_data.columns
```

```python
# Impute Categorical value with most frequent one
Imputed_Data_mode
```

```python
Original_df = pd.merge(Imputed_Data_mean, Imputed_Data_mode, how = 'outer', left_index = True, right_index = True)
```
