```python
numerical_data = games.drop('Year', axis=1).select_dtypes(include='float64')
```

```python
numerical_data
```

```python
imputer = IterativeImputer(max_iter=10, random_state=0)
```

```python
imputed_data = pd.DataFrame(imputer.fit_transform(numerical_data))
```

```python
imputed_data.columns = numerical_data.columns 
```

```python
imputed_data.isnull().sum()
```
