```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('D:\karan data\Data for Preprocessing\Movie_classification.csv')
```

```python
df.head()
```

```python
df.info()
```

### Missing Value Imputation

```python
df['Time_taken'].mean()
```

```python
df['Time_taken'].fillna(value = df['Time_taken'].mean(), inplace = True)
```

```python
df.info()
```

### Dummy Variable Creation

```python
df.head()
```

```python
df = pd.get_dummies(df,columns = ["3D_available","Genre"],drop_first = True)
```

```python
df.columns
```

```python
change_type = ['3D_available_YES', 'Genre_Comedy','Genre_Drama', 'Genre_Thriller']
```

```python
for i in change_type:
  df[i] =  df[i].astype(int)
  
```

```python
df.head()
```

### X-y split

```python
X = df.loc[:,df.columns!="Start_Tech_Oscar"]
type(X)
```

```python
X.head()
```

```python
X.shape
```

```python
y = df["Start_Tech_Oscar"]
type(y)
```

```python
y.head()
```

```python
y.shape
```

### Test-Train Split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.2,random_state=0)
```

```python
X_train.head()
```

```python
X_train.shape
```

```python
X_test.shape
```

### Training Classification Tree

```python
from sklearn import tree
clftree = tree.DecisionTreeClassifier(max_depth = 3)
```

```python
clftree.fit(X_train, y_train)
```

### Predict values using trained model

```python
y_train_pred = clftree.predict(X_train)
y_test_pred = clftree.predict(X_test)
```

```python
y_test_pred
```

### Model Performance

```python
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
confusion_matrix(y_train, y_train_pred)
```

```python
confusion_matrix(y_test, y_test_pred)
```

```python
accuracy_score(y_test, y_test_pred)
```

### Plotting decision tree

```python
feature_names_list = list(X_train.columns)
plt.figure(figsize=(40,12))
tree.plot_tree(clftree, filled=True, feature_names=feature_names_list, fontsize=20)
plt.show()
```

## Controlling Tree growth

```python
clftree2 = tree.DecisionTreeClassifier(min_samples_leaf = 20, max_depth=4)
clftree2.fit(X_train, y_train)
feature_names_list = list(X_train.columns)
plt.figure(figsize=(40,12))
tree.plot_tree(clftree, filled=True, feature_names=feature_names_list, fontsize=30)
plt.show()
```

```python
accuracy_score(y_test, clftree2.predict(X_test))
```

### Grid Search

https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.GridSearchCV.html

```python
from sklearn.model_selection import GridSearchCV
```

```python
rf_clf = RandomForestClassifier(n_estimators=250, random_state=42)
```

```python
params_grid = {'max_features' : [4,5,6,7,8,9,10],
               'min_samples_split': [2,3,10]
               }
```

```python
grid_search = GridSearchCV(rf_clf, params_grid, n_jobs = -1, cv=5, scoring = 'accuracy')
```

```python
grid_search.fit(X_train, y_train)
```

```python
grid_search.best_params_ 
```

```python
cvrf_clf = grid_search.best_estimator_
```

```python
accuracy_score(y_test, cvrf_clf.predict(X_test))
```

```python
confusion_matrix(y_test, cvrf_clf.predict(X_test))
```
