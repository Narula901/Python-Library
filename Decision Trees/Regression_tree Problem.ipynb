### Missing Value Imputation

```python
df['Time_taken'].isnull().sum()
```

```python
df['Time_taken'].mean()
```

```python
df['Time_taken'].fillna(value = df['Time_taken'].mean(), inplace = True)
```

```python
df.info()
```

### Dummy Variable Creation

```python
df.head()
```

```python
df = pd.get_dummies(df, columns=['3D_available', 'Genre'], drop_first = True).astype(int)
```

```python
df.head()
```

### X-y split

```python
X = df.loc[:, df.columns!='Collection']
```

```python
X.head()
```

```python
X.shape
```

```python
y = df['Collection']
type(y)
```

```python
y.head()
```

```python
y.shape
```

### Test-Train Split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
X_train.head()
```

```python
X_train.shape
```

```python
X_test.shape
```

### Traing Regression Tree

```python
from sklearn import tree
```

```python
regtree = tree.DecisionTreeRegressor(max_depth=3)
```

```python
regtree.fit(X_train, y_train)
```

```python
feature_name = list(X_train.columns)
```

### Predict values using trained model

```python
y_train_pred = regtree.predict(X_train)
y_test_pred = regtree.predict(X_test)
```

```python
y_test_pred
```

### Model Performance

```python
from sklearn.metrics import mean_squared_error,r2_score
```

```python
mean_squared_error(y_test, y_test_pred)
```

```python
r2_score(y_train, y_train_pred)
```

```python
r2_score(y_test, y_test_pred)
```

### Plotting decision tree

```python
# Plot the tree
plt.figure(figsize=(40,8))
tree.plot_tree(regtree, filled=True, feature_names=feature_name)
plt.show()
```

```python
X_train.columns
```

## Controlling Tree growth

### Maximum number of levels in tree

```python
regtree = tree.DecisionTreeRegressor(max_depth=3)
regtree.fit(X_train, y_train)
plt.figure(figsize=(40,8))
tree.plot_tree(regtree, filled=True, feature_names=feature_name)
plt.show()
```

### Minimum observations at internal node

```python
regtree = tree.DecisionTreeRegressor(min_samples_split=40)
regtree.fit(X_train, y_train)
plt.figure(figsize=(40,8))
tree.plot_tree(regtree, filled=True, feature_names=feature_name)
plt.show()
```

### Minimum observations at leaf node

```python
regtree = tree.DecisionTreeRegressor(min_samples_leaf=25, max_depth = 4)
regtree.fit(X_train, y_train)
plt.figure(figsize=(40,8))
tree.plot_tree(regtree, filled=True, feature_names=feature_name, fontsize=15)
plt.show()
```
